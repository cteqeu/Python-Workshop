## Python: List Comprehension

```python
uitprinten = []
x = ['hallo','hoe','is','het']
for i in x:
    uitprinten.append(str(i).upper())
print(uitprinten)
```

```python
uitprinten2 = []
getallen = [ 77, 88, 99]
for q in getallen:
    uitprinten2.append(q**3)
print(uitprinten2)
```

```python
[z**3 for z in getallen]
```

```python
[str(v).upper() for v in x]
```

## Python - Map and Filter

```python
sequentie = range(7)
```

```python
map(maal2,sequentie)
```

```python
list(map(maal2,sequentie))
```

```python
list(map(lambda variabele: variabele*2, sequentie))
```

```python
filter(lambda item: item%3 == 0 ,sequentie)
```

```python
list(filter(lambda item:item%3 == 0, sequentie))
```
